```python
credit = pd.read_csv("./homework2.csv",na_values="?")
def c12Aproval(x):
    if (x=="+"):
        return 1
    return 0

credit["Approved"] = credit.C1.map(c12Aproval)
#credit.drop(["C1"])
credit
```

```python
from numpy.random import randn

def makeColor(x):
    if x:
        return 'green'
    return 'red'

credit['color'] = credit.Approved.map(makeColor)
  
```

```python
#to do - fill in na
#create variables for the other objects

def fillWithMean(x):
    _mask = credit[x].isnull()
    mean = credit[x].mean()
    std = credit[x].mean()
    credit.loc[_mask,x] = np.random.normal(mean,std)
    
numerics = ["A2","A3","A8","A11","A14","A15"]
for x in numerics:
    fillWithMean(x)
credit.info()
```

```python
#credit.A1.unique
categories = ["A1","A4","A5","A6","A9","A10","A12","A13"]
def createCat(colname):
    a = pd.unique(credit[colname].values.ravel())
    i = 1
    lookup = {}
    for x in a:
        lookup[x] = i
        i = i + 1
    credit[colname+"val"] = credit[colname].map(lambda x: lookup[x] + np.random.normal(0,0.1))
 
for x in categories:   
    createCat(x)
credit.info()
```

```python
#at least they have the same parameters
#let's steal some more from the class

# We generate a grid in the square [-3,3 ]^2.
xx, yy = np.meshgrid(np.linspace(0, 100, 500),
                     np.linspace(0, 100, 500))

# This function takes a SVM estimator as input.
def plot_decision_function(fitted_est,X,y,title):
    fig = plt.figure();
    ax = fig.add_subplot(1, 1, 1)
    ax.scatter(X[:, 0],X[:, 1],c=.5+.5*y,s=70)
    ax.set_xlim([0,X[:, 0].max()*1.1])
    ax.set_ylim([0,X[:, 1].max()*1.1])
    #line stuff
    Z = fitted_est.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cmap = plt.cm.coolwarm
    # We display the decision function on the grid.
    #plt.figure(figsize=(5,5));
    plt.imshow(Z,
                extent=(xx.min(), xx.max(), yy.min(), yy.max()),
                aspect='auto', origin='lower', cmap=cmap);
    plt.contour(xx, yy, Z, levels=[0], linewidths=2,colors='k');
    ax.set_title(title)
    
#yes we should plot everything rather than the test    
label = "Run 1 " + str(est.score(x_t,y_t))
plot_decision_function(est.best_estimator_,x_t,y_t,label)
```
