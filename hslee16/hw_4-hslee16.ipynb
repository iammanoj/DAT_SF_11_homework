```python
# Standard imports for data analysis packages in Python
import pandas as pd
import numpy as np
import seaborn as sns  # for pretty layout of plots
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import pdist, squareform
from sklearn import preprocessing
from blaze import *

# This enables inline Plots
%matplotlib inline
```

```python
# Read data and assign category column
wine = pd.read_csv('../data/wine.csv')
y = wine.Wine

# Drop the first column for KMeans
wine.drop(labels='Wine', axis=1, inplace=True)
```

```python
#looks like phenols and flavanoids columns have a strong correlation
pd.scatter_matrix(wine, figsize=(13, 13), diagonal='kde')
```

```python
# train using kmeans clustering and fit the data
kmeans = KMeans(n_clusters=3, init='random', n_init=10 , max_iter = 300, random_state=1)
y_hat_kmeans = kmeans.fit(wine).labels_
```

```python
y_hat_kmeans
```

```python
# the labels do not match
y.values
```

```python
plt.scatter(wine.Alcohol, wine['Malic.acid'], c=y)
```

```python
# Even plotting the clusters shows that the clusters are similar but not close to exact
plt.scatter(wine.Alcohol, wine['Malic.acid'], c=y_hat_kmeans)
```

```python
plt.scatter(wine.Phenols, wine.Flavanoids, c=y_hat_kmeans)
```

```python
#using other columns, we can observe that the clusters output by kmeans are also similar 
plt.scatter(wine.Phenols, wine.Flavanoids, c=y)
```

```python
# using PCA 
pca = PCA()
```

```python
X_pca = pca.fit_transform(wine)
```

```python
pd.scatter_matrix(pd.DataFrame(X_pca), figsize=(10, 10), diagonal='kde')
```

```python
print np.cov(X_pca, rowvar=False)
```

```python
variance =  pca.explained_variance_ratio_

for i in variance:
    print i
```

```python
# lets try scaling the data first
pca = PCA()
scaled = preprocessing.scale(wine)
X_pca = pca.fit_transform(scaled)
```

```python
pca.explained_variance_ratio_
```

```python
# looks like the variance ratio is more spread across the columns when the data is initially scale
plt.scatter(X_pca[:, 0], X_pca[:, 1])
```

```python
kmeans = KMeans(n_clusters=3, init='random', n_init=10 , max_iter = 300, random_state=1)
y_hat_kmeans = kmeans.fit(X_pca).labels_

#by scaling the features first, we get a much more distinct set of  clusters
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_hat_kmeans)
```

```python
# plot centroids
mu = kmeans.cluster_centers_
plt.scatter(X_pca[:,0], X_pca[:,1], c=y_hat_kmeans, alpha=0.4)
plt.scatter(mu[:,0], mu[:,1], s=100, c=np.unique(y_hat_kmeans))
```

```python
# It looks like the first column (alcohol) is responsible for 99.8% of the variance when not scaled.
# Lets try retraining the model with just one component 
pca = PCA(1)
X_pca = pca.fit_transform(wine)
```

```python
pca.components_
```

```python
pca.mean_
```

```python
variance =  pca.explained_variance_ratio_

for i in variance:
    print i
```

```python
# Using PCA with just 1 component results in the same variance as the first column (alcohol)
# shown above. 
```

```python
# lets train kmeans with the reduced dimensions to see if the same clusters as we observed above are present
pca = PCA(5)
X_pca = pca.fit_transform(wine)

# Use the output of PCA in kmeans
kmeans = KMeans(n_clusters=3, init='random', n_init=10 , max_iter = 300, random_state=1)
y_hat_kmeans = kmeans.fit(X_pca).labels_
```

```python
data = pd.DataFrame(X_pca)
```

```python
# Let's try hierarchical clustering
dist_wine = squareform(pdist(scaled, metric='euclidean'))
```

```python
# Looks like the single method produce a large set of clusters
R = dendrogram(linkage(dist_wine, method='single'), color_threshold=10)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```

```python
# The ward method produces more distinctcluster sets 
R = dendrogram(linkage(dist_wine, method='ward'), color_threshold=20)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```
